```python
import pandas as pd

# Creamos exactamente la misma tabla de la profe
datos_profe = {
    'ciudad': ['Buenos Aires', 'Córdoba', 'Mendoza', 'Rosario', 'Tucumán', 'Buenos Aires', 'Córdoba'],
    'monto': [1000000, 800000, 1432110, 1711890, 1598230, 823450, 854320]
}

df = pd.DataFrame(datos_profe)

# Vemos cómo queda la tabla
print(df)
```

```python
df = dataframe = pd.DataFrame(datos_profe)
print(df.head())
print(df.dtypes)
```

```python
grupos = df.groupby("ciudad")
print(type(grupos))
# <class 'pandas.core.groupby.DataFrameGroupBy'>
```

```python
print("--- TOTAL DE VENTAS (.sum()) ---")
print(df.groupby("ciudad")["monto"].sum())

print("\n--- PROMEDIO DE VENTAS (.mean()) ---")
print(df.groupby("ciudad")["monto"].mean())

print("\n--- CANTIDAD DE VENTAS (.count()) ---")
print(df.groupby("ciudad")["monto"].count())
```

```python
import pandas as pd

# Creamos la tabla con ciudad, categoría y monto
datos_profe = {
    'ciudad': ['Buenos Aires', 'Buenos Aires', 'Córdoba', 'Córdoba', 'Mendoza', 'Mendoza'],
    'categoria': ['Electrónica', 'Ropa', 'Electrónica', 'Ropa', 'Electrónica', 'Ropa'],
    'monto': [1200000, 623450, 1000000, 654320, 900000, 532110]
}

df = pd.DataFrame(datos_profe)

# Ejecutamos el agrupamiento doble
print(df.groupby(["ciudad", "categoria"])["monto"].sum())
```

```python
import pandas as pd

# Datos con la columna 'vendedor'
datos_vendedores = {
    'vendedor': ['Ana', 'Ana', 'Carlos', 'Carlos', 'Ana', 'Carlos'],
    'monto': [150000, 200000, 100000, 300000, 50000, 250000]
}

df = pd.DataFrame(datos_vendedores)

# Aplicamos .agg() con la lista de funciones
resumen = df.groupby("vendedor")["monto"].agg(["sum", "mean", "count", "max"])
print(resumen)
```

```python
import pandas as pd

# Creamos el dataset completo con todas las columnas que pide el apunte
datos = {
    'ciudad': ['Buenos Aires', 'Buenos Aires', 'Córdoba', 'Córdoba'],
    'monto': [1000, 2000, 1500, 2500],
    'unidades': [2, 5, 3, 4],
    'descuento': [10, 15, 5, 20]
}

df = pd.DataFrame(datos)

# Ejecutamos el resumen profesional
resumen_pro = df.groupby("ciudad").agg(
    total_ventas   = ("monto",    "sum"),
    ticket_promedio= ("monto",    "mean"),
    unidades_total = ("unidades", "sum"),
    descuento_max  = ("descuento","max"),
)

print(resumen_pro)
```
